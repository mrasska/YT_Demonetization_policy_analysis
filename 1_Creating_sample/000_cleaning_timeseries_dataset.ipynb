```python
import pandas as pd
import numpy as np
```

```python
db_timeseries=pd.read_csv('df_timeseries_en.tsv', sep='\t')
```

```python
db_timeseries.info()
```

```python
db_timeseries['datetime']=pd.to_datetime(db_timeseries['datetime'], yearfirst=True)
db_timeseries['week'] = db_timeseries['datetime'].apply(lambda x: x.strftime('%Y-%U')) 
```

```python
db_timeseries.head()
```

## Keeping info between 1st January 2016 and 30th December 2018

```python
db_deepdive=db_timeseries[db_timeseries['week'].between('2016-01', '2018-52')]
```

```python
db_deepdive.head()
```
