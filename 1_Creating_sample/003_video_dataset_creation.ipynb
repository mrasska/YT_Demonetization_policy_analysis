### Information regarding the code
This script regroups the partition of the video dataset. Some columns from the chunks are dropped as the data will not be exploited in the study.

```python
import pandas as pd
import numpy as np
```

```python
d0=pd.read_csv('video_dataset_chunk_0.csv', sep=';')
d0=d0.drop(['Unnamed: 0','dislikes', 'like', 'view'], axis=1)

d0.info()
```

```python
d0['channel'].nunique()
```

```python
d0['upload']=pd.to_datetime(d0['upload'], yearfirst=True)

# Dropping rows for videos published before 4th January 2016
d0 = d0[d0['upload']>='2016-01-04']

# Dropping rows for videos published after 31st September 2019
d0 = d0[d0['upload']<'2017-12-31']
```

```python
d0.info()
```

```python
d0['channel'].nunique()
```
