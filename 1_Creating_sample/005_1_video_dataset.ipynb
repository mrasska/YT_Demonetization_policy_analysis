### Regrouping videos per channel per week

```python
db_video['upload']=pd.to_datetime(db_video['upload'], yearfirst=True)
db_video['week'] = db_video['upload'].apply(lambda x: x.strftime('%Y-%W')) 
```

```python
db_video.drop_duplicates(subset=['id'])
db_video.info()
```

```python
db_grouped=db_video.groupby(['channel', 'week', 'before_updates', 'category'], 
                            as_index=False, 
                           dropna=False)['id'].count()
db_grouped
```

```python
db_grouped['week']=db_grouped['week'].replace('2017-00', '2016-52')
```

```python
db_grouped=db_grouped.groupby(['channel', 'week', 'before_updates', 'category'], 
                            as_index=False, 
                           dropna=False)['id'].sum()
db_grouped
```
