## Channel categorization
In this notebook, we labelled YouTube channels based on the most popular video genre. We focused on the pre-intervention period for the labelling and defined the most popular genre as the one with a share equal or larger than 51%.

```python
import pandas as pd
import numpy as np
```

```python
db_week=pd.read_csv('weekly_uploads_cat_completed.csv', sep=';')
db_week.info()
```

```python
condition=[(db_week["week"]>="2017-12"),
          (db_week["week"]<"2017-12")]
value=[1,0]
db_week["first_moderation_update"]=np.select(condition, value)
```

```python
db_weekly=db_week.groupby(['channel', 'week', 'first_moderation_update', 
                           'cat_people', 'cat_ent', 'cat_educ', 'cat_comedy', 
                           'cat_how', 'cat_auto', 'cat_travel', 'cat_game',
                           'cat_science', 'cat_activ', 'cat_pet', 'cat_film', 
                           'cat_music', 'cat_news', 'cat_sport', 'cat_unknown'],
                           as_index=False, 
                           dropna=False)['id_0'].sum(min_count=1)
db_weekly
```

```python
db_weekly=db_weekly[db_weekly['first_moderation_update']==0]

db_chan=db_weekly.groupby(['channel'],
                           as_index=False, 
                           dropna=False)['id_0', 'cat_people', 'cat_ent', 'cat_educ', 'cat_comedy', 
                           'cat_how', 'cat_auto', 'cat_travel', 'cat_game',
                           'cat_science', 'cat_activ', 'cat_pet', 'cat_film', 
                           'cat_music', 'cat_news', 'cat_sport', 'cat_unknown'].sum(min_count=1)
db_chan
```

```python
db_chan['people_rate']=(db_chan['cat_people']/db_chan['id_0'])*100
db_chan['ent_rate']=(db_chan['cat_ent']/db_chan['id_0'])*100
db_chan['educ_rate']=(db_chan['cat_educ']/db_chan['id_0'])*100
db_chan['comedy_rate']=(db_chan['cat_comedy']/db_chan['id_0'])*100
db_chan['how_rate']=(db_chan['cat_how']/db_chan['id_0'])*100
db_chan['auto_rate']=(db_chan['cat_auto']/db_chan['id_0'])*100
db_chan['travel_rate']=(db_chan['cat_travel']/db_chan['id_0'])*100
db_chan['game_rate']=(db_chan['cat_game']/db_chan['id_0'])*100
db_chan['science_rate']=(db_chan['cat_science']/db_chan['id_0'])*100
db_chan['activ_rate']=(db_chan['cat_activ']/db_chan['id_0'])*100
db_chan['pet_rate']=(db_chan['cat_pet']/db_chan['id_0'])*100
db_chan['film_rate']=(db_chan['cat_film']/db_chan['id_0'])*100
db_chan['music_rate']=(db_chan['cat_music']/db_chan['id_0'])*100
db_chan['news_rate']=(db_chan['cat_news']/db_chan['id_0'])*100
db_chan['sport_rate']=(db_chan['cat_sport']/db_chan['id_0'])*100
db_chan['unknown_rate']=(db_chan['cat_unknown']/db_chan['id_0'])*100


db_chan
```

```python
condition=[(db_chan['people_rate']>=51), (db_chan['ent_rate']>=51),
          (db_chan['educ_rate']>=51), (db_chan['comedy_rate']>=51),
          (db_chan['how_rate']>=51), (db_chan['auto_rate']>=51),
          (db_chan['travel_rate']>=51), (db_chan['game_rate']>=51),
          (db_chan['science_rate']>=51), (db_chan['activ_rate']>=51),
          (db_chan['pet_rate']>=51), (db_chan['film_rate']>=51),
          (db_chan['music_rate']>=51), (db_chan['news_rate']>=51),
          (db_chan['sport_rate']>=51), (db_chan['unknown_rate']>=51)]

value=['People', 'Entertainment', 'Education', 'Comedy', 'How to & Style', 'Autos & Vehicles', 
      'Travel', 'Gaming', 'Science & Technology', 'Nonprofits & Activism', 'Pets & Animals', 
      'Film & Animation', 'Music', 'News & Politics', 'Sports', 'Unknown']

db_chan['main_cat']=np.select(condition, value)
```

```python
db_chan
```

```python
db_chan['main_cat']=db_chan['main_cat'].replace('0', 'Mixed')
```

```python
db_chan_cat=db_chan[['channel', 'main_cat']].copy()
```

```python
db_chan_cat
```

```python
db_chan_cat.to_csv('channel_cat.csv', sep=';')
```
