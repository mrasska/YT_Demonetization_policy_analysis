## Additional channel selection

As mentionned in our paper, we looked into channels whose main video genre is not associated with traditionnal media (films and animation, music, news and politics, and sports). For our analysis, we removed channels linked to these topics

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 100)
```

```python
db_week=pd.read_csv('weekly_uploads_cat_completed.csv', sep=';')
db_week.head()
```

```python
db=db_week[['week', 'channel', 'id_0']].copy()
db.info()
```

```python
db=db.groupby(['channel', 'week'], as_index=False)['id_0'].sum()
db
```

```python
db['week']=db['week'].replace('2017-00', '2016-52')
```

```python
db=db.groupby(['channel', 'week'], as_index=False)['id_0'].sum()
db
```

```python
db['week'].nunique()
```

### Removing channels not in greenlighted categories

```python
db_chan_cat=pd.read_csv('channel_cat.csv', sep=';')
db_chan_cat
```

```python
db=pd.merge(db, db_chan_cat, how='left', left_on='channel', right_on='channel')
db
```

```python
db=db[(db['main_cat']!='Music') & (db['main_cat']!='Film & Animation')
      & (db['main_cat']!='News & Politics') & (db['main_cat']!='Sports')  
      & (db['main_cat']!='Unknown')]
db
```

```python
db=db.drop(['Unnamed: 0'], axis=1)
```

```python
db['channel'].nunique()
```

```python
condition=[(db["week"]>="2017-12"),
          (db["week"]<"2017-12")]
value=[0, 1]
db["before_update"]=np.select(condition, value)
db
```

### removing channels not active prior moderation update

```python
db_before=db[db['before_update']==1]
```

```python
test=db_before.groupby('channel', as_index=False)['id_0'].sum()
test
```

```python
test=test[test['id_0']==0]
test
```

```python
unique_before=test['channel'].unique()
len(unique_before)
```

```python
unique_total=db['channel'].unique()
len(unique_total)
```

```python
db=db[~db["channel"].isin(unique_before)]
db
```

```python
db['channel'].nunique()
```

```python
db['week'].nunique()
```

```python
max(db['id_0'])
```

### Channel's average audience size before moderation update

We calculate the average number of followers for each channel. This enables us to estimate our model with the subscribers number brackets.

```python
db_timeseries=pd.read_csv("timeseries_cleaned.csv", sep=';')
db_timeseries
```

```python
db_weekly=pd.merge(db, db_timeseries, how='left', left_on=['channel', 'week'], right_on=['channel', 'week'])
```

```python
db_weekly
```

```python
max(db_weekly['id_0'])
```

```python
db_average=db_weekly[db_weekly['before_update']==1]
db_average
```

```python
db_average=db_average.groupby('channel', as_index=False, dropna=False)['subs'].mean().round(0)
db_average
```

```python
final=pd.merge(db, db_average, how='left', left_on='channel', right_on='channel')
final
```

```python
max(final['id_0'])
```

```python
test=final[final['subs'].isna()]
test
```

```python
test['channel'].nunique()
```

```python
final=final[~final['subs'].isna()]
final
```

```python
final.to_csv('weekly_uploads_cat_completed.csv', sep=';')
```
