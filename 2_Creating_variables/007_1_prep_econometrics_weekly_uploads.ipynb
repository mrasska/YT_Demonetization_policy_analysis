```python
import pandas as pd
import numpy as np
```

## Regrouping data per weeks

```python
db_video=pd.read_csv('weekly_uploads_cat_completed.csv', sep=';')
db_video.info()
```

```python
# videos published after 1st content moderation update
condition=[(db_video["week"]>="2017-12"),
          (db_video["week"]<"2017-12")]
value=[1, 0]
db_video["first_moderation_update"]=np.select(condition, value)
```

```python
db_video
```

```python
db_video['week']=db_video['week'].replace('2017-00', '2016-52')
```

```python
db_video
```

```python
db_weekly=db_video.groupby(['channel', 'week', 
                           'first_moderation_update'],
                           as_index=False, 
                           dropna=False)['id_0'].sum(min_count=1)
db_weekly
```

```python
db_weekly[db_weekly['week']=='2016-52']
```

#### Adding channels informations

```python
db_treat=pd.read_csv('id_channels_treatment_2024.csv', sep=';')
db_treat
```

```python
db_weekly=pd.merge(db_weekly, db_treat, how='left', left_on='channel', right_on='id_channel')
db_weekly
```

```python
db_weekly=db_weekly.drop(['Unnamed: 0', 'id_channel'],
            axis=1)
```

```python
db_out=pd.read_csv('outliers_995.csv', sep=';')
db_out
```

```python
db_weekly=db_weekly[~db_weekly['channel'].isin(db_out['channel'])]
db_weekly
```

```python
db_weekly
```

```python
condition=[(db_weekly['type']=='treatment')]

value=['treatment']
db_weekly['type']=np.select(condition, value)
db_weekly
```

```python
db_weekly['type'].value_counts()
```

```python
db_weekly['type']=db_weekly['type'].replace('0', 'control')
db_weekly
```

```python
db_weekly['type'].value_counts()
```

```python
db_chan_cat=pd.read_csv('channel_cat.csv', sep=';')
db_chan_cat
```

```python
db_weekly=pd.merge(db_weekly, db_chan_cat, how='left', left_on='channel', right_on='channel')
db_weekly
```

```python
db_timeseries=pd.read_csv("timeseries_cleaned.csv", sep=';')
db_timeseries
```

```python
db_weekly=pd.merge(db_weekly, db_timeseries, how='left', left_on=['channel', 'week'], right_on=['channel', 'week'])
```

```python
db_weekly
```

```python
db_weekly=db_weekly.drop(['Unnamed: 0_x', 'Unnamed: 0_y', 'views', 'delta_views', 'delta_subs', 'videos',
                         'delta_videos', 'activity'],
            axis=1)
```
