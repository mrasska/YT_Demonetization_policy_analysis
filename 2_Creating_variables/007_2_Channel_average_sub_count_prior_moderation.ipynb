```python
import pandas as pd
import numpy as np
```

```python
db_week=pd.read_csv('true_weekly_uploads_umbalanced_data.csv', sep=';')
db_week.head()
```

```python
db_week.info()
```

```python
db_before=db_week[db_week['first_moderation_update']==0]
db_before
```

```python
db_channel_average=db_before.groupby(['channel'],
                           as_index=False, 
                           dropna=False)['subs'].mean().round()

db_channel_average
```

```python
min_sub=min(db_channel_average['subs'])
max_sub=max(db_channel_average['subs'])

bins=[min_sub,10000, 100000, 1000000, max_sub]
labels=['Less than 10k subs','Between 10k and 100k subs','Between 100k and 1M subs', 'More than 1M subs']
db_channel_average['sub_bin_prior']=pd.cut(x=db_channel_average['subs'], bins=bins, labels=labels, include_lowest=True)
db_channel_average.head()
```

```python
db_channel_average.to_csv('channel_average_subs_prior_mod.csv', sep=';')
```

```python
db_week['channel'].nunique()
```

```python
db_week=pd.merge(db_week, db_channel_average, how='left', left_on='channel', right_on='channel')
db_week
```
